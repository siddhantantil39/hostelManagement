```python
df = pd.read_csv("hostel.csv")
```

```python

last_row = df.iloc[-1]
#last_row = df[df.USN == '1SI19CS410']
v = np.array(last_row)[0]
USN = np.array(last_row)[0]
v
```

```python
df['rating']= 0
```

```python
df = df[:-1]
```

```python
df
```

```python
for i in df.index :
    if df['DOB'][i]== int(last_row.DOB)  : 
         df.loc[i,'rating'] =  int(df.loc[i,'rating']) + 1
    if df['STATE'][i]== v[3]: 
         df.loc[i,'rating'] =  int(df.loc[i,'rating']) + 1
    if df['BRANCH'][i]== v[4]: 
        df.loc[i,'rating'] =  int(df.loc[i,'rating']) + 1
```

```python
df
```

```python
df_pivot=df.pivot_table(columns='ROOM NO.',index='USN',values='rating')
```

```python
df_pivot.fillna(3,inplace=True)
```

```python
df_pivot
```

```python
from scipy.sparse import csr_matrix
df_sparse=csr_matrix(df_pivot)
```

```python
from sklearn.neighbors import NearestNeighbors
model=NearestNeighbors( n_neighbors=7,algorithm='brute',metric='cosine')
```

```python
model.fit(df_sparse)
```

```python
distances,suggestions=model.kneighbors(df_pivot.iloc[0,:].values.reshape(1,-1))
```
